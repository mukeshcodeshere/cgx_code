```python
import requests
key = "USEYOURKEY HERE"
def getdata (symbol, market, start_time, end_time, key = key):
    url = "https://cryptocurrency-financial-data.p.rapidapi.com/trades"

    querystring = {
        "base": "USDT",
        "market_venue": market,
        "symbol": symbol,
        "start": start_time,
        "end": end_time,
        "limit": "50000",  # You can adjust this limit as per your requirement
        "sort": "asc"
    }

    headers = {
        "Accept": "application/json",
        "X-RapidAPI-Key": "0b99f4e31cmsh8f055536f05fe87p1243b9jsn43792cb5a9e1",
        "X-RapidAPI-Host": "cryptocurrency-financial-data.p.rapidapi.com"
    }

    response = requests.get(url, headers=headers, params=querystring)

    # Check if the request was successful
    if response.status_code == 200:
        data = response.json()["data"]
        # Check if data is not empty
        if data:
            # Convert JSON data to DataFrame
            df = pd.json_normalize(data)
            df.to_csv(f'{symbol}_{market}_trade_data.csv', index = False)
            # Print DataFrame
            print(df)
        else:
            print("No data returned.")
    else:
        print("Error:", response.status_code)


```
