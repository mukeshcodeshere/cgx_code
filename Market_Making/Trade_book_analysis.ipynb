```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
exchanges = ['BINANCE', 'BYBIT', 'OKX', 'KUCOIN', 'BITSTAMP', 'COINBASE', 'KRAKEN']
df_exchanges = {}
for exchange in exchanges:  
    df = pd.read_csv(f'BTC_{exchange}_trade_data.csv')
    df['timestamp'] = pd.to_datetime(df['timestamp'], format='ISO8601')
    df_exchanges[exchange] = df
```

```python
for exchange in df_exchanges:
    plt.plot(df_exchanges[exchange]['timestamp'], df_exchanges[exchange]['price'],label = exchange)
plt.legend
# we can get rid of the exchanges that do not have enough data
```

```python
# Aggregate volume
volume = []
for exchange in exchanges:
    volume.append(df_exchanges[exchange]['volume'].sum())
# we will use the fair price as the volume weighted average last price among the 4 most traded exchanges updating every 0.l seconds
```

```python
plt.pie(volume, labels= exchanges)
```
